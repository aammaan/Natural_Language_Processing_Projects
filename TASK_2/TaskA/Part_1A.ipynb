```python
import json
from sklearn.model_selection import train_test_split

with open('NER_TRAIN_JUDGEMENT.json') as f:
    data = json.load(f)

train_data, val_data = train_test_split(data, test_size=0.15, random_state=42)
```

```python
import re

def custom_tokenizer(text):
    # Define regular expression pattern to tokenize words and special characters
    pattern = r'\w+|[^\w\s]'
    tokens = re.findall(pattern, text)
    return tokens

# Example usage:
text = "This is an example sentence with, some special characters!"
tokens = custom_tokenizer(text)
print(tokens)
```

```python
clean_data = {}
label_map=set()
label_map.add('O')
global max_seq_len
max_seq_len=0

for entry in train_data:
    id = entry['id']
    text = entry['data']['text']
    tokens = custom_tokenizer(text)
    token_label = ['O']*len(tokens)
    max_seq_len=max(max_seq_len,len(tokens))
    result_list = entry['annotations'][0]['result']
    for val in result_list:
        vtext = val['value']['text']
        val_label = val['value']['labels'][0]
        vtokens = custom_tokenizer(vtext)
        for vidx, vtoken in enumerate(vtokens):
            for idx, token in enumerate(tokens):
                if vtoken==token:
                    if vidx == 0:
                        token_label[idx] = 'B_' + val_label
                        label_map.add('B_'+val_label)

                    else:
                        if token_label[idx-1] != 'O':
                            token_label[idx] = 'I_' + val_label
                            label_map.add('I_'+val_label)

    clean_data[id] = {}
    clean_data[id]['text'] = tokens
    clean_data[id]['labels'] = token_label
print(max_seq_len)

with open('NER_train.json', 'w') as f:
    json.dump(clean_data, f, indent=4)
```
